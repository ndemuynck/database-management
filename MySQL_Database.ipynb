### Take data from a Pandas DataFrame to create a new data table

```python
user = 'user'
pwd = 'pwd'
host = 'db4free.net'
schema = 'pokemon1'  # this is our database name

cnxn = create_engine(
    f"mysql+mysqlconnector://{user}:{pwd}@{host}/{schema}",
    echo=False)
```

```python
# name = new table name
dataset.to_sql(name='pokemon', con=cnxn, if_exists='append', index=False)
```
